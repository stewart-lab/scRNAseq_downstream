# Pseudotime 

## First we use palantir to construct a pseudo time

```python
# Palantir
# source venv/bin/activate
# load packages

import palantir
import scanpy as sc
import pandas as pd
import os

# Plotting
import matplotlib
import matplotlib.pyplot as plt

# warnings
import warnings
from numba.core.errors import NumbaDeprecationWarning

warnings.filterwarnings(action="ignore", category=NumbaDeprecationWarning)
warnings.filterwarnings(
    action="ignore", module="scanpy", message="No data for colormapping"
)

# Inline plotting
%matplotlib inline

# read config file
import json
GIT_DIR = '/w5home/bmoore/scRNAseq_downstream/'
with open(GIT_DIR+'config.json') as f:
    config_dict = json.load(f)
    print(config_dict)

# variables
DATA_DIR = config_dict["pseudotime"]["DATA_DIR"]
ADATA_FILE = config_dict["pseudotime"]["ADATA_FILE"] 
ANNOT_TYPE =  config_dict["pseudotime"]["ANNOT_TYPE"] #"manual", "seurat_map", "clustifyr", "manual", "orig.ident"
CROSS_SPECIES = config_dict["pseudotime"]["CROSS_SPECIES"] # "TRUE" or "FALSE"
NC = config_dict["pseudotime"]["NC"] # number of components

```

```python
# load data
## note- data was previously converted from seurat object to anndata object

data_dir = os.path.expanduser(DATA_DIR)
adata = sc.read_h5ad(data_dir + ADATA_FILE)
adata
from datetime import datetime
now = datetime.now()
now = now.strftime("%Y%m%d_%H%M%S")
out_dir = data_dir + "pseudotime_" + now +"/"
os.mkdir(out_dir)
```

```python
adata
with open(GIT_DIR+'config.json') as f:
    config_dict = json.load(f)
    print(config_dict)
```

```python
# check metadata
annot_label = config_dict["pseudotime"]["annot_label"]
# # check indices
# adata_top = adata.obs.head()
# list(adata_top.index)
# # read in metadata
# metadata = pd.read_csv(data_dir + "manual_annot_metadata.txt", index_col=0, sep="\t")
# # add metadata
# adata.obs['CellType1'] = metadata['CellType1']
# check if added
print(adata.obs[annot_label].unique())
print(adata.obs.columns)
```

```python
# visualize (note we have already done nearest neighbor and umap)
with plt.rc_context():
    sc.pl.embedding(
        adata,
        basis="umap",
        frameon=False,
        color=annot_label,
        )
    plt.savefig(out_dir + "labeled_clusters.pdf")
```

```python

## gene expression visualization on UMAP
# get inputs from config
genes = config_dict["pseudotime"]["gene_list"]
magic_impute = config_dict["pseudotime"]["magic_impute"]

# MAGIC imputation
# Palantir can use MAGIC to impute the data for visualization and determining gene expression trends.
if magic_impute == "TRUE":
    imputed_X = palantir.utils.run_magic_imputation(adata)
    # umap with magic
    with plt.rc_context():
        sc.pl.embedding(
        adata,
        basis="umap",
        layer="MAGIC_imputed_data",
        color=genes,
        frameon=False,
        show=False,
        )
    plt.savefig(out_dir + "gene_expression_magic.pdf")
else:
    print("no magic imputation, showing true gene expression only")

# umap with gene expression
with plt.rc_context():
    sc.pl.embedding(
    adata,
    basis="umap",
    use_raw=False,
    color=genes,
    frameon=False,
    show=False,
    )
    plt.savefig(out_dir + "gene_expression.pdf")
```

```python
# Run diffusion maps
dm_res = palantir.utils.run_diffusion_maps(adata, n_components=NC)
ms_data = palantir.utils.determine_multiscale_space(adata)
```

```python
# Diffusion maps visualization
palantir.plot.plot_diffusion_components(adata)
plt.savefig(out_dir + 'palantir_components_umap.pdf')
plt.show()
```

```python
# find terminal states
terminal_celltypes = config_dict["pseudotime"]["terminal_celltypes"]
terminal_states = palantir.utils.find_terminal_states(adata, celltypes=terminal_celltypes, celltype_column=annot_label)
print(terminal_states)
```

```python
# find start cell
start_celltype = config_dict["pseudotime"]["start_celltype"]
start_cell = palantir.utils.early_cell(adata,celltype="Retinal Prog", celltype_column=annot_label, 
                                       fallback_seed=1234)
```

```python
# use cells found for start cell and terminal states for palantir analysis

# Run Palantir
if len(terminal_states) == 0:
    lista = [start_cell]
    pr_res = palantir.core.run_palantir(adata, start_cell, num_waypoints=500, use_early_cell_as_start=True)
    # plot with start cell
    palantir.plot.highlight_cells_on_umap(adata, lista)
    plt.savefig(out_dir + 'palantir_start_cell.pdf')
    plt.show()
else:
    pr_res = palantir.core.run_palantir(
    adata, start_cell, num_waypoints=500, terminal_states=terminal_states
    )
    # plot with terminal cells and start cell
    all_cells = terminal_states.copy()
    all_cells[start_cell] = "Start cell"
    palantir.plot.highlight_cells_on_umap(adata, all_cells)
    plt.savefig(out_dir + 'palantir_terminal_cells.pdf')
    plt.show()
    print(all_cells)
```

```python
# plot results
palantir.plot.plot_palantir_results(adata, s=3)
plt.savefig(out_dir + 'palantir_results.pdf')
plt.show()
```

```python
# Terminal state probability distributions of individual cells can be visualized using the plot_terminal_state_probs function

if len(terminal_states) == 0:
    lista = [start_cell]
    palantir.plot.plot_terminal_state_probs(adata, lista)
    plt.savefig(out_dir + 'terminal_state_probs.pdf')
    plt.show()
else:
    cells=all_cells.index
    print(cells)
    palantir.plot.plot_terminal_state_probs(adata, cells)
    plt.tight_layout()
    plt.savefig(out_dir +'terminal_state_probs.pdf', bbox_inches='tight')
    plt.show()
```

```python
# Branch probabilities

# Select cells
# Before computing the gene expression trends, we first need to select cells associated with a specific branch of the pseudotime trajectory.
masks = palantir.presults.select_branch_cells(adata, eps=0)
# visualize the branck selection
palantir.plot.plot_branch_selection(adata)
plt.savefig(out_dir + 'branch_selection.png')
plt.show()
```

```python
gene_trends = palantir.presults.compute_gene_trends(
    adata, expression_key="MAGIC_imputed_data",)
print(adata)
#gene_trend_key
```

```python
# Determine marker gene trends using the  Mellon Function Estimator
gene_trends = palantir.presults.compute_gene_trends(
    adata,
    expression_key="MAGIC_imputed_data",
)
# visualize trends
if CROSS_SPECIES == "FALSE":
    if ANNOT_TYPE == "orig.ident":
        genes = ["MIXL1","HES1","HES5","HES7","LFNG","DLL1","MSGN1"]
        palantir.plot.plot_gene_trends(adata, genes)
        plt.savefig(out_dir + 'gene_trends.png')
        plt.show()
        
    else:
        genes = ["CNGA3", "RCVRN","CA10", "TUBB3","SOX2"]
        palantir.plot.plot_gene_trends(adata, genes)
        plt.savefig(out_dir + 'gene_trends.png')
        plt.show()
else:
    genes = ["CNGA3", "RCVRN", "CA10"]
    palantir.plot.plot_gene_trends(adata, genes)
    plt.savefig(out_dir + 'gene_trends.png')
    plt.show()
    
```

```python
# Heatmap trend visualization
palantir.plot.plot_gene_trend_heatmaps(adata, genes)
plt.savefig(out_dir + 'gene_trends_heatmap.png')
plt.show()
```

```python
# Clustering
# Gene expression trends can be clustered and visualized
# Here the first 1000 genes along the cones genes are clustered
more_genes = adata.var_names[:1000]
if ANNOT_TYPE == "clustifyr":
    # clustifyr
    communities = palantir.presults.cluster_gene_trends(adata, "Cones", more_genes)
    palantir.plot.plot_gene_trend_clusters(adata, "Cones")
elif ANNOT_TYPE == "manual":
    # manual annotation
    communities = palantir.presults.cluster_gene_trends(adata, 'Cones', more_genes)
    palantir.plot.plot_gene_trend_clusters(adata, 'Cones')
elif ANNOT_TYPE == "seurat_map":
    # seurat mapping
    communities = palantir.presults.cluster_gene_trends(adata, "PR", more_genes)
    palantir.plot.plot_gene_trend_clusters(adata, "PR")
elif ANNOT_TYPE == "orig.ident":
    # timepoint/original annotation
    communities = palantir.presults.cluster_gene_trends(adata, "ResyncTime5", more_genes)
    palantir.plot.plot_gene_trend_clusters(adata, "ResyncTime5")
else:
    print("Error: invalid annotation type")
# save figure and show
plt.savefig(out_dir + 'gene_trend_clusters.png')
plt.show()
```

```python
import numpy as np
import cellrank as cr
import scanpy as sc
import scvelo as scv
# compute diffusion pseudotime (DPT)
# start by computing a diffusion map
# need to recompute nearest neighbots first
sc.pp.neighbors(adata, n_pcs=30, n_neighbors=30, random_state=0)
sc.tl.diffmap(adata)
```

```python
# find root cell
root_ixs = adata.obsm['X_diffmap'][:, 3].argmax()
print(root_ixs)
```

```python
# visualize and set root

if ANNOT_TYPE == "clustifyr":
    scv.pl.scatter(
    adata,
    basis="diffmap",
    c=['clustifyr_call_consol_type', root_ixs],
    legend_loc="right",
    components=["2, 3"],
    save= out_dir +'diffusion_map.png'
    )
    
elif ANNOT_TYPE == "manual":
    scv.pl.scatter(
    adata,
    basis="diffmap",
    c=['CellType_manual', root_ixs],
    legend_loc="right",
    components=["2, 3"],
    save= out_dir +'diffusion_map.png'
    )
    
elif ANNOT_TYPE == "seurat_map":
    scv.pl.scatter(
        adata,
        basis="diffmap",
        c=['predicted.celltype', root_ixs],
        legend_loc="right",
        components=["2, 3"],
        save= out_dir +'diffusion_map.png'
    )
elif ANNOT_TYPE == "orig.ident":
    scv.pl.scatter(
        adata,
        basis="diffmap",
        c=['orig.ident', 62], 
        legend_loc="right",
        components=["2, 3"],
        save= out_dir +'diffusion_map.png'
    )
    
adata.uns["iroot"] = root_ixs
```

```python
# Compute DPT and compare it with the precomputed Palantir pseudotime:
sc.tl.dpt(adata)
with plt.rc_context():
    sc.pl.embedding(
    adata,
    basis="umap",
    color=["dpt_pseudotime", "palantir_pseudotime"],
    color_map="gnuplot2",
    show=False
    )
    plt.savefig(out_dir +'dpt-palantir_pseudotime.png')
```

```python
# plot trajectory
cone_trajectory = "NA"
bipolar_trajectory = "NA"
rod_trajectory = "NA"
pr_trajectory = "NA"
img_trajectory = "NA"
ac_trajectory = "NA"
time_trajectory = "NA"
ganglion_trajectory = "NA"
mg_trajectory = "NA"
all_trajectory = "NA"

if ANNOT_TYPE == "clustifyr":
    cone_trajectory = ["Retinal Prog", "Pan PRs", "Cones"]
    bipolar_trajectory = ["Retinal Prog", "Muller Glia", "Bipolar Cells"]
    rod_trajectory = ["Retinal Prog", "Pan PRs"]
    key= "clustifyr_call_consol_type"
elif ANNOT_TYPE == "manual":
    cone_trajectory = ['Retinal Prog', 'unknown', 'Cones']
    ganglion_trajectory = ['Retinal Prog', 'unknown', 'Ganglion Cells']
    all_trajectory = ['Retinal Prog', 'Ganglion Cells', 'Cones']
    key= "CellType_manual"
elif ANNOT_TYPE == "seurat_map":
    pr_trajectory = ["Prog","Prog/Glia","PR"]
    bipolar_trajectory = ["Prog","Prog/Glia","BC"]
    img_trajectory = ["Prog","Prog/Glia","iMG"]
    ac_trajectory = ["Prog","Prog/Glia","AC"]
    key= "predicted.celltype"
elif ANNOT_TYPE == "orig.ident":
    time_trajectory = ["ResyncTime1","ResyncTime2","ResyncTime3","ResyncTime4","ResyncTime5","ResyncTime6"]
    key= "orig.ident"

# # plot the cone trajectory
if cone_trajectory != "NA":
    mask = np.in1d(adata.obs[key], cone_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=cone_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'cone_trajectory.png', bbox_inches='tight')

# # plot the bipolar trajectory
if bipolar_trajectory != "NA":
    mask = np.in1d(adata.obs[key], bipolar_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=bipolar_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'bipolar_trajectory.png', bbox_inches='tight')

# # plot the rod trajectory
if rod_trajectory != "NA":
    mask = np.in1d(adata.obs[key], rod_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=rod_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'rod_trajectory.png', bbox_inches='tight')

if ganglion_trajectory != "NA":
    mask = np.in1d(adata.obs[key], ganglion_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=ganglion_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'ganglion_trajectory.png', bbox_inches='tight')

if mg_trajectory != "NA":
    mask = np.in1d(adata.obs[key], mg_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=mg_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'mg_trajectory.png', bbox_inches='tight')

# plot the PR trajectory
if pr_trajectory != "NA":
    mask = np.in1d(adata.obs[key], pr_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=pr_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'pr_trajectory.png', bbox_inches='tight')

# plot the iMG trajectory
if img_trajectory != "NA":
    mask = np.in1d(adata.obs[key], img_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=img_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'img_trajectory.png', bbox_inches='tight')

# plot the AC trajectory
if ac_trajectory != "NA":
    mask = np.in1d(adata.obs[key], ac_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=ac_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'ac_trajectory.png', bbox_inches='tight')
# plot the time trajectory
if time_trajectory != "NA":
    mask = np.in1d(adata.obs[key], time_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=time_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'time_trajectory.png', bbox_inches='tight')

if all_trajectory != "NA":
    mask = np.in1d(adata.obs[key], all_trajectory)
    with plt.rc_context():
        sc.pl.violin(
        adata[mask],
        keys=["dpt_pseudotime", "palantir_pseudotime"],
        groupby=key,
        rotation=-90,
        order=all_trajectory,
        show=False
        )
        plt.tight_layout()
        plt.savefig(out_dir +'all_trajectory.pdf', bbox_inches='tight')
```
